```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import font_manager, rc # 폰트 세팅을 위한 모듈 추가
```

```python
font_path = "C:/Windows/Fonts/malgun.ttf" # 사용할 폰트명 경로 삽입
font = font_manager.FontProperties(fname = font_path).get_name()
rc('font', family = font)

```

```python
def getWeighted(name,weightConst):
    forestdf = pd.read_csv(name+'.csv')
    forestdf.dropna()
    goo = forestdf.columns[0]
    event = forestdf.columns[1]
    forestdf = forestdf.sort_values(by=[goo], axis=0)


    dfsum = forestdf[event].sum()
    forestdf[event] = forestdf[event].apply( lambda x: ((x/dfsum)+1)**weightConst)

    
    return forestdf
```

```python
def plotDF(df):
    names = df[df.columns[0]]
    values = df[df.columns[1]]
    
    plt.bar(range(16), values, tick_label=names)
    plt.xticks(rotation='vertical')
    plt.show()
```

```python
gooSubter = getWeighted('구군별지하면적',1)
plotDF(gooSubter)
```

```python
gooSubter = getWeighted('부산 농경지',1)
gooSubter
```

```python
gooSubter = getWeighted('재난CCTV',1)
plotDF(gooSubter)

```

```python
gooSubter = getWeighted('해안선_원본',1)
plotDF(gooSubter)
```
