```python
import pathlib

import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from collections import defaultdict

import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

d = defaultdict(LabelEncoder)
print(tf.__version__)
```

```python
goose = ['금정구','동래구','해운대구','기장군','남구','동구','중구','부산진구','영도구','사상구','사하구','수영구','북구','연제구','강서구']
```

```python
def small_df(name):
    sdf = pd.read_csv(name+".csv")
    sdf.insert(2, "LandName",name, True)
    return sdf
```

```python
initial = small_df(goose[0])

for k in goose[1:]:
    initial = pd.concat([initial,small_df(k)], ignore_index=True)
df = initial
```

```python

replaceDic ={'총강수량(mm)': 'total_rain','강수계속시간(시간)':'rain_duration','최다강수1H':'max_rain',
                   '일강수량(mm)':'daily_rain','누적 강수량(mm)-2일':'culmu_rain_2','누적 강수량(mm)-3일':'culmu_rain_3',
                   '최대풍속(m/s)':'max_wind','평균기온(℃)':'avg_temp','기간(일)':'duration' ,'침수면적(1,000㎡)':'sunk','태풍':'typhoon','피해액(백만원)':'damage'}
df.rename(columns = replaceDic,inplace = True)
```

```python
blacklist = ['max_rain','max_wind','daily_rain','damage','culmu_rain_2','culmu_rain_3','total_rain']

df[blacklist] = df[blacklist].fillna(0)
df['avg_temp'] = df['avg_temp'].fillna(df['avg_temp'].mean())
df = df.dropna()
```

```python
df['damage'] =df['damage'].apply(lambda x:float(str(x).replace(",","")))
def classify(num):
    if(num > 600000):
        return 4
    if(100000< num < 500000):
        return 3
    if(50000 < num < 100000):
        return 2
    if(10000 < num < 50000):
        return 1
    return 0
df['damage'] =df['damage'].apply(classify )
```

```python
df
```

```python
amount_data = ['damage','duration','total_rain','max_wind']
```

```python
landf = pd.get_dummies(df['LandName'],dtype = int)
typhoondf = pd.get_dummies(df['typhoon'],dtype = int)
my_land =landf.columns.values.tolist()
my_typhoon = typhoondf.columns.values.tolist()
df[my_land] = landf
df[my_typhoon] = typhoondf
```

```python
train_dataset = df[amount_data+my_land+my_typhoon].sample(frac=0.8,random_state=0)
test_dataset = df[amount_data+my_land+my_typhoon].drop(train_dataset.index)
test_dataset
```

```python
train_stats = train_dataset.describe()
train_stats.pop("damage")
train_stats = train_stats.transpose()
train_stats
```

```python
train_labels = train_dataset.pop('damage')
test_labels = test_dataset.pop('damage')
```

```python
def norm(x):
  return (x - train_stats['mean']) / train_stats['std']
normed_train_data = norm(train_dataset)
normed_test_data = norm(test_dataset)
normed_train_data
```

```python

def build_model():
  model = keras.Sequential([
    layers.Dense(32, activation='relu', input_shape=[len(train_dataset.keys())]),
    layers.Dense(1)
  ])

  optimizer = tf.keras.optimizers.Adam(0.001)

  model.compile("adam", "binary_crossentropy", metrics=["accuracy"])

  return model
```

```python
model = build_model()
```

```python
EPOCHS = 1000
class PrintDot(keras.callbacks.Callback):
  def on_epoch_end(self, epoch, logs):
    if epoch % 100 == 0: print('')
    print('.', end='')
history = model.fit(
  normed_train_data, train_labels,
  epochs=EPOCHS, validation_split = 0.2, verbose=0,
  callbacks=[PrintDot()])
```

```python
hist = pd.DataFrame(history.history)
hist['epoch'] = history.epoch

print(hist.round(2))
```

```python
realnorm=norm(df[amount_data[1:]+my_land+my_typhoon])
realnorm
```

```python

test_predictions = model.predict(realnorm).flatten()
damagelist = df.loc[test_dataset.index.values.tolist()]['damage']
print(damagelist.max())
for idx,k in enumerate(damagelist):
    print("real=",k,"prediction=",test_predictions[idx])
print(len(test_predictions))
```
