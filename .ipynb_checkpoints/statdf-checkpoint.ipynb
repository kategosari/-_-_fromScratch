```python
from pandas import DataFrame
import pandas as pd
goodic= {'강서구': 195,
 '금정구': 143,
 '기장군': 265,
 '남구': 140,
 '동구': 123,
 '동래구': 47,
 '부산진구': 271,
 '북구': 271,
 '사상구': 165,
 '사하구': 71,
 '서구': 167,
 '수영구': 288,
 '연제구': 146,
 '영도구': 200,
 '중구': 20,
 '해운대구': 211}

names = list(goodic.keys())
values = list(map(lambda x:x/100,goodic.values()))
data = {'구':names,'위험도':values}
data_df = DataFrame(data) 
data_df.to_csv('구별위험도.csv', index=False, encoding="utf-8-sig")
```
